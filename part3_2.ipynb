```python
def show_video(path='video'):
  # This function loads the data video inline into the colab notebook.
  # By reading the video stored by the Monitor class.
  mp4list = Path(path).glob('*.mp4')
  mp4list = sorted(mp4list)
  html = []
  if len(mp4list) > 0:
    for mp4 in mp4list:
      encoded = base64.b64encode(mp4.read_bytes())
      html.append('''<video alt="{}" autoplay loop controls 
                    style="height: 256px;">
                    <source src="data:video/mp4;base64,{}" type="video/mp4" />
             </video>'''.format(mp4, encoded.decode('ascii')))
    ipythondisplay.display(ipythondisplay.HTML(data="<br>".join(html)))
  else: 
    print("Could not find video")
```

```python
!pip install git+https://github.com/eleurent/rl-agents
from rl_agents.agents.common.factory import agent_factory
```

```python
env = gym.make("highway-v0")

env = gym.wrappers.Monitor(env, './video', force=True, 
                           video_callable=lambda episode: True)

agent_config = {
    "__class__": "<class 'rl_agents.agents.tree_search.deterministic.DeterministicPlannerAgent'>",
    "env_preprocessors": [{"method":"simplify"}],
    "budget": 100,
    "gamma": 0.70
}

agent = agent_factory(env, agent_config)

obs, done = env.reset(), False
env.unwrapped.automatic_rendering_callback = env.video_recorder.capture_frame

for step in trange(env.unwrapped.config["duration"], desc="Running Lane Changing Task Episodes"):
    action = agent.act(obs)
    obs, reward, done, info = env.step(action)

env.close()
del agent
del agent_config
```
